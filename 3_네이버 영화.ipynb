### 6. 네이버 영화 평점 기준 영화의 평점 변화 확인하기

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20200331'

html = urlopen(url)

soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
soup.find_all('div','tit5')
```

```python
# 제목
soup.find_all('div','tit5')[0].a.string
```

```python
# 평점
soup.find_all('td','point')[0].string
```

```python
date = pd.date_range('2019-12-20', periods=100, freq='D')
date
```

```python
# 100일간의 영화 평점 및 랭킹 수집
import urllib

movie_date = []
movie_title = []
movie_score = []

for today in tqdm_notebook(date):
    html = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}'
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, 'html.parser')
    end = len(soup.find_all('td','point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_title.extend([soup.find_all('div','tit5')[n].a.string for n in range(0,end)])
    movie_score.extend([soup.find_all('td','point')[n].string for n in range(0,end)])
```

```python
movie = pd.DataFrame({'date':movie_date,'title':movie_title, 'score':movie_score})
movie['score'] = movie['score'].astype(float)
movie.head()
```

```python
# 영화별 점수 합산
import numpy as np

movie_unique = pd.pivot_table(movie, index=['title'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='score',ascending=False)
movie_best.head()
```

```python
# 특정 영화만 추려서 확인하기
tmp = movie.query('title==["결혼 이야기"]') # 따옴표에 주의
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'],tmp['score'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
